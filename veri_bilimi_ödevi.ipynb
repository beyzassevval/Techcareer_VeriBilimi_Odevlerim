```python
import numpy as np

#1.soru
array1=np.random.randint(0,100,(5,5))
print(array1)
#ortalama,standart sapma ve varyans hesabı
ortalama=np.mean(array1)
std_dev = np.std(array1)
varyans = np.var(array1)
print("Ortalama:",ortalama)
print("Standart Sapma:", std_dev)
print("Varyans:", varyans)

# En büyük ve en küçük değerleri bulma
en_buyuk = np.max(array1)
en_kucuk = np.min(array1)
print("En Büyük Değer:", en_buyuk)
print("En Küçük Değer:", en_kucuk)

# Köşegen elemanlarının toplamını bulma
kosegen_toplam = np.trace(array1)
print("Köşegen Elemanları Toplamı:", kosegen_toplam)
```

```python
#2. Numpy ile Veri Simülasyonu
import numpy as np

mu, sigma = 50, 20 # normal dağılım için
array2 = np.random.normal(mu, sigma, 1000)
array2 = np.clip(array2, 0, 100)

print(array2)
print("--"*50)

# Ortalama, medyan ve standart sapmayı hesaplayın.
ortalama=np.mean(array2)
print("Ortalama:",ortalama)

std_sapma=np.std(array2)
print("Standart sapma:",std_sapma)

medyan=np.median(array2)
print("Medyan:",medyan)

# 50’den düşük alan kaç öğrenci olduğunu bulun.
low_scores_count = np.sum(array2 < 50)
print("50'den düşük alan öğrenci sayısı:", low_scores_count)
```

```python
import pandas as pd
from random import randint


data={"Öğrenci":['Ali','Ayşe','Can','Damla','Fatma','Nur','Şevval','Beyza','Bahadır','Yiğit'],
      "Yaş":[randint(18,30) for i in range(10)],
      "Bölüm":["Bilgisayar","Endüstri","Elektrik","Makine","Bilgisayar","Endüstri","Elektrik","Makine","Elektrik","Bilgisayar"],
      "Matematik":[randint(0,100) for i in range(10)],
      "Fizik":[randint(0,100) for i in range(10)],
      "Kimya":[randint(0,100) for i in range(10)]}

  # Veri setini Pandas DataFrame olarak oluşturun.
df = pd.DataFrame(data)
display(df)
print("--"*50)
# her ders için ortalama puan
print(f"Matematik dersinin ortalaması: {df['Matematik'].mean()}")
print(f"Fizik dersinin ortalaması: {df['Fizik'].mean()}")
print(f"Kimya dersinin ortalaması: {df['Kimya'].mean()}")
# En yüksek matematik notu
print(f"En yüksek matematik notu: {df['Matematik'].max()}")
```

```python
📘 Veri Bilimi Ödevi – Numpy ve Pandas Uygulamaları
1. Numpy Bölümü
Aşağıdaki adımları uygulayın:
1.	1. Numpy ile Matris İşlemleri
- 5x5 boyutunda rastgele (0-100 arasında) tam sayılardan oluşan bir matris oluşturun.
- Bu matrisin:
  • Ortalama, standart sapma ve varyansını hesaplayın.
  • En büyük ve en küçük değerlerini bulun.
  • Köşegenindeki elemanların toplamını bulun.
2.	2. Numpy ile Veri Simülasyonu
- 1000 adet öğrencinin sınav puanlarını (0-100 arası, normal dağılımdan gelen) simüle edin.
- Ortalama, medyan ve standart sapmayı hesaplayın.
- 50’den düşük alan kaç öğrenci olduğunu bulun.
2. Pandas Bölümü
Bir CSV dosyası oluşturun ya da aşağıdaki gibi örnek bir DataFrame hazırlayın:
Öğrenci	Yaş	Bölüm	Matematik	Fizik	Kimya
Ali	20	Bilgisayar	70	65	80
Ayşe	21	Fizik	60	75	85
Mehmet	19	Kimya	80	70	65
Zeynep	22	Bilgisayar	90	85	95
Ahmet	20	Fizik	55	60	70
•	Görevler:
3.	1. Veri setini Pandas DataFrame olarak oluşturun.
4.	2. Her ders için ortalama puanı bulun.
5.	3. En yüksek matematik notunu alan öğrenciyi bulun.
6.	4. Her öğrencinin not ortalamasını hesaplayan yeni bir sütun ekleyin.
7.	5. Bölümlere göre gruplayarak her bölümün ortalama başarılarını hesaplayın.
8.	6. Ortalaması 70’in üzerinde olan öğrencileri filtreleyin.
```

```python
# Her öğrencinin not ortalamasını hesaplayan yeni bir sütun
df['Ortalama Not'] = df[['Matematik', 'Fizik', 'Kimya']].mean(axis=1)
display(df)
# Bölümlere göre gruplayalım,her bölümün not ortalaması
print("##"*50)
print("##"*50)
bolume_gore_basari = df.groupby('Bölüm')['Ortalama Not'].mean()
print("Bölümlere Göre Ortalama Başarı:")
display(bolume_gore_basari)
print("##"*50)
print("##"*50)
#ortalaması 70 üzerinde olanlar
print("Ortalaması 70'in üzerinde olanlar:")
display(df[df['Ortalama Not'] > 70])
```
