```python
import numpy as np

#1.soru
array1=np.random.randint(0,100,(5,5))
print(array1)
#ortalama,standart sapma ve varyans hesabÄ±
ortalama=np.mean(array1)
std_dev = np.std(array1)
varyans = np.var(array1)
print("Ortalama:",ortalama)
print("Standart Sapma:", std_dev)
print("Varyans:", varyans)

# En bÃ¼yÃ¼k ve en kÃ¼Ã§Ã¼k deÄŸerleri bulma
en_buyuk = np.max(array1)
en_kucuk = np.min(array1)
print("En BÃ¼yÃ¼k DeÄŸer:", en_buyuk)
print("En KÃ¼Ã§Ã¼k DeÄŸer:", en_kucuk)

# KÃ¶ÅŸegen elemanlarÄ±nÄ±n toplamÄ±nÄ± bulma
kosegen_toplam = np.trace(array1)
print("KÃ¶ÅŸegen ElemanlarÄ± ToplamÄ±:", kosegen_toplam)
```

```python
#2. Numpy ile Veri SimÃ¼lasyonu
import numpy as np

mu, sigma = 50, 20 # normal daÄŸÄ±lÄ±m iÃ§in
array2 = np.random.normal(mu, sigma, 1000)
array2 = np.clip(array2, 0, 100)

print(array2)
print("--"*50)

# Ortalama, medyan ve standart sapmayÄ± hesaplayÄ±n.
ortalama=np.mean(array2)
print("Ortalama:",ortalama)

std_sapma=np.std(array2)
print("Standart sapma:",std_sapma)

medyan=np.median(array2)
print("Medyan:",medyan)

# 50â€™den dÃ¼ÅŸÃ¼k alan kaÃ§ Ã¶ÄŸrenci olduÄŸunu bulun.
low_scores_count = np.sum(array2 < 50)
print("50'den dÃ¼ÅŸÃ¼k alan Ã¶ÄŸrenci sayÄ±sÄ±:", low_scores_count)
```
