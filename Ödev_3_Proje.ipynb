```python
#Proje – “Kitap Satış Analiz Sistemi”

#Veriler
kitaplar = [
    {"isim": "Veri Bilimi 101", "yazar": "Ali", "tur": "Bilim", "satis": 1200, "yil": 2021},
    {"isim": "Python ile Yapay Zeka", "yazar": "Ayşe", "tur": "Bilim", "satis": 950, "yil": 2020},
    {"isim": "İstatistik Temelleri", "yazar": "Ali", "tur": "Akademik", "satis": 700, "yil": 2019},
    {"isim": "Makine Öğrenmesi", "yazar": "Can", "tur": "Bilim", "satis": 1800, "yil": 2022},
    {"isim": "Veri Görselleştirme", "yazar": "Deniz", "tur": "Sanat", "satis": 400, "yil": 2018},
    {"isim": "Matematiksel Modelleme", "yazar": "Ali", "tur": "Akademik", "satis": 1500, "yil": 2021},
    {"isim": "Bilgi Toplumu", "yazar": "Ayşe", "tur": "Sosyal", "satis": 600, "yil": 2022},
    {"isim": "Yapay Zeka ve Etik", "yazar": "Deniz", "tur": "Sosyal", "satis": 820, "yil": 2023},
    {"isim": "Algoritmaların Gücü", "yazar": "Can", "tur": "Bilim", "satis": 1300, "yil": 2021},
    {"isim": "Sanat ve Veri", "yazar": "Ece", "tur": "Sanat", "satis": 540, "yil": 2020},
    {"isim": "Veri Madenciliği", "yazar": "Fatma", "tur": "Bilim", "satis": 1100, "yil": 2019},
    {"isim": "Toplum ve Teknoloji", "yazar": "Gökhan", "tur": "Sosyal", "satis": 770, "yil": 2021},
    {"isim": "Görsel İstatistik", "yazar": "Ece", "tur": "Akademik", "satis": 650, "yil": 2018},
    {"isim": "Kuantum Hesaplama", "yazar": "Ali", "tur": "Bilim", "satis": 900, "yil": 2022},
    {"isim": "Dijital Dönüşüm", "yazar": "Ayşe", "tur": "Sosyal", "satis": 1400, "yil": 2023},
    {"isim": "Sanatta Yapay Zeka", "yazar": "Deniz", "tur": "Sanat", "satis": 480, "yil": 2019},
    {"isim": "Veri Analizi Uygulamaları", "yazar": "Can", "tur": "Bilim", "satis": 1250, "yil": 2020},
    {"isim": "Matematiğin Estetiği", "yazar": "Fatma", "tur": "Akademik", "satis": 890, "yil": 2021},
    {"isim": "Kültür ve Teknoloji", "yazar": "Gökhan", "tur": "Sosyal", "satis": 560, "yil": 2019},
    {"isim": "Bilim Tarihi", "yazar": "Ece", "tur": "Bilim", "satis": 730, "yil": 2018}
]
# En çok satanlar
en_cok_satilanlar = sorted(kitaplar, key=lambda x: x["satis"], reverse=True)
print("En çok satilanlar:",en_cok_satilanlar)
#yazar_satislari(kitaplar)
def yazar_satislari(kitaplar):
  satislar = {}
  for kitap in kitaplar:
    yazar = kitap["yazar"]
    satis = kitap["satis"]
    if yazar in satislar:
      satislar[yazar] += satis
    else:
      satislar[yazar] = satis
  return satislar
  yazarlar = yazar_satislari(kitaplar)
print("Yazarlar",yazarlar)

#en çok satılan yazar
en_cok_satilan_yazar = max(yazarlar, key=yazarlar.get)
print(f"En çok satılan yazar: {en_cok_satilan_yazar}")
```

```python
def kitap_türleri(kitaplar):
  tur_satislari = {}
  for kitap in kitaplar:
    tur = kitap["tur"]
    satis = kitap["satis"]
    if tur in tur_satislari:
      tur_satislari[tur] += satis
    else:
      tur_satislari[tur] = satis
  return tur_satislari

tur_satislari = kitap_türleri(kitaplar) # Call the function and assign the result
print("Türler",tur_satislari)
#en çok satılan tür
en_cok_satilan_tur = max(tur_satislari, key=tur_satislari.get)
print(f"En çok satılan tür: {en_cok_satilan_tur}")
print("--"*50)
#satış adeti 1000 den fazla olanları listeleyelim
cok_satan_kitaplar = []
for kitap in kitaplar:
  if kitap["satis"] > 1000:
    cok_satan_kitaplar.append(kitap)
print("--"*50)
print("Satış adeti 1000'den fazla olan kitaplar:")
print("--"*50)
for kitap in cok_satan_kitaplar:
  print(f"- {kitap['isim']} ({kitap['satis']} satış)")
  print("--"*50)
```

```python
#Lambda / Filter / Map Kullanımı:

kitaplar_2020_sonrasi = list(filter(lambda kitap: kitap["yil"] >= 2020, kitaplar))
print("2020 sonrası çıkan kitaplar:", kitaplar_2020_sonrasi)
print("--"*50)
satislar_artirilmis = list(map(lambda kitap: kitap["satis"] * 1.1, kitaplar))
print("Satış adetleri %10 artırılmış:", satislar_artirilmis)
print("--"*50)
kitaplar.sort(key=lambda kitap: kitap["satis"], reverse=True)
print("Kitaplar satış miktarına göre azalan şekilde sıralanmış:")
print("--"*50)
for kitap in kitaplar:
  print(kitap)
```

```python
import statistics

# Ortalama Satış adedi
toplam_satis = sum(kitap["satis"] for kitap in kitaplar)
ortalama_satis = toplam_satis / len(kitaplar)
print("Ortalama Satış Miktarı:", ortalama_satis)
#en çok satan ürün
en_cok_satilan_urun=max(kitaplar,key=lambda x:x["satis"])
print("En Çok Satan Ürün:",en_cok_satilan_urun)
#standart sapma
standart_sapma = statistics.stdev(kitap["satis"] for kitap in kitaplar)
print("Satış Standart Sapma:", standart_sapma)
```

```python
import pandas as pd

df = pd.DataFrame(kitaplar)

X = df[['yazar', 'tur', 'yil']]
y = df['satis']

print("Features (X):")
display(X.head())
print("\nTarget (y):")
display(y.head())
```

```python
from sklearn.model_selection import train_test_split
xTrain,xTest,yTrain,yTest=train_test_split(X,y,test_size=0.3,random_state=42)

print(xTrain.shape)
print(xTest.shape)
print(yTrain.shape)
print(yTest.shape)
```

```python
# Train de yazarların ortalama satışı
yazar_satis_train = {}
for index, row in xTrain.iterrows():
    yazar = row['yazar']
    satis = yTrain[index]
    if yazar in yazar_satis_train:
        yazar_satis_train[yazar].append(satis)
    else:
        yazar_satis_train[yazar] = [satis]

yazar_ortalama_satis = {yazar: sum(satis_list) / len(satis_list) for yazar, satis_list in yazar_satis_train.items()}

print("Yazarların Ortalama Satışları:")
print(yazar_ortalama_satis)
#test verisindeki ortalama üstü satışlar
test_ort_üst={}
for index, row in xTest.iterrows():
    yazar = row['yazar']
    satis = yTest[index]
    if yazar in test_ort_üst:
        test_ort_üst[yazar].append(satis)
    else:
        test_ort_üst[yazar] = [satis]

test_ort_üst = {yazar: sum(satis_list) / len(satis_list) for yazar, satis_list in test_ort_üst.items()}
print("--"*50)
print("test verisindeki ortalama üstü satışlar:")
print(test_ort_üst)
```
